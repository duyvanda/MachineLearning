```python
# Function for polynomial display
# We do not need pay much attention on this part
def plot_polynomial(xmin, xmax, coef, color='C1'):
    #xs is an array of evenly spaced numbers between xmin and xmax
    xs = np.linspace(xmin, xmax, num=500)
    #ys is an array, each element is computed as a polynomial function of
    #the corresponding element of xs
    ys = np.zeros_like(xs)
    for p, c in enumerate(coef.flatten()):
        ys += c*np.power(xs, p)
    plt.plot(xs, ys, color=color)
```

```python
# See how it fits:
plt.scatter(X0, Y)
plot_polynomial(0, 10, theta)
plt.show()
```

```python
def dJtheta(X, Y, theta_gd, m):
    """
    X: X's value
    Y: Y's value
    theta_gd: theta's value
    m: number of samples
    """
    dJtheta = 1/m * X.T.dot(X.dot(theta_gd) - Y)
    return dJtheta.reshape(theta_gd.shape)
```

```python
def loss(X, Y, theta_gd, m):
    """
    X: X's value
    Y: Y's value
    theta_gd: theta's value
    m: number of samples
    """
    Y_predicted = X.dot(theta_gd)
    return 1/(2*m) * np.sum((Y_predicted - Y)**2)
```

```python
theta_new = np.array([[0.],
       [ 0.]])
```

```python
m = X.shape[0]
```

```python
learning_rate = 0.02
```

```python
for i in range(10000):
    theta_new = theta_new - learning_rate*dJtheta(X, Y, theta_new, m)
    print(loss(X, Y, theta_new, m))
    print ("loss {}, m {}, b {} iteration {}".format(loss(X, Y, theta_new, m),theta_new[0],theta_new[1], i))
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.02

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {} iteration {}".format(m_curr,b_curr,cost, i))

x = np.array([[2], [7], [9], [3], [10], [6], [1], [8]])
y = np.array([[13], [35], [41], [19], [45], [28], [10], [55]])

gradient_descent(x,y)
```
